## Question 2

### What is a scatter plot?
#### A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.title("Scatter plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.scatter(x, y)
```

## Question 3

### Why is the subplot() function used?
#### subplot() function adds subplot to a current figure at the specified grid position. The subplot() function takes three arguments that describes the layout of the figure. The layout is organized in rows and columns, which are represented by the first and second argument. The third argument represents the index of the current plot. #the figure has 1 row, 2 columns, and this plot is the first plot.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 3, 1)
plt.plot(x, y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 3, 2)
plt.plot(x, y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 3, 3)
plt.plot(x, y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 3, 4)
plt.plot(x, y)
plt.show()
```

## Question 4

### What is a bar plot? Why is it used?
#### Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data. The Matplotlib library in Python is widely used to create bar plots.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.title("Company Growth")
plt.xlabel("Companies")
plt.ylabel("Profits")
plt.bar(company, profit)
```

```python
plt.barh(company, profit)
```

## Question 5

### What is a box plot? Why is it used?
#### A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2])
plt.show()
```

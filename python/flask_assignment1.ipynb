### What is Flask Framework?
#### Flask is a web framework, it’s a Python module that lets developer develop web applications easily. It’s has a small and easy-to-extend core: it’s a microframework that doesn’t include an ORM (Object Relational Manager) or such features. It does have many cool features like url routing, template engine. It is a WSGI web app framework.

### What are the advantages of Flask Framework?
1. Scalable. Size is everything, and Flask's status as a microframework means that we can use it to grow a tech project such as a web app incredibly quickly.
2. Flexible.
3. Easy to negotiate.
4. Lightweight.
5. Documentation.
6. Not a lot of tools.
7. Difficult to get familiar with a larger Flask app.
8. Less maintenance costs.

## Question 3

### What is App routing in Flask? Why do we use app routes?

#### App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. 
#### To bind a function to an URL path we use the app.route decorator. In above application(example), the URL (“/”) is associated with the root URL. So if the site’s domain was https://mango-pharmacist-stlgy.pwskills.app and we want to add routing to “https://mango-pharmacist-stlgy.pwskills.app:5000/hello”, we would use “/hello” along with the port number.

## Question 5

### What function is used in Flask for URL Building?
#### The url_for() function is used in Flask for URL building.
#### In the below example the library management system which can be used by the two types of users, admin and student. There is a specific function named user() which recognizes the user then redirect the user to the exact function which contains the implementation for this particular function.

```
from flask import Flask
   
app = Flask(__name__)  
  
@app.route('/admin')  
def admin():  
    return 'admin' 
  
@app.route('/student')  
def student():  
    return 'student'  
  
@app.route('/user/<name>')  
def user(name):  
    if name == 'admin':  
        return redirect(url_for('admin'))
    if name == 'student':  
        return redirect(url_for('student'))

if __name__=="__main__":
    app.run(host="0.0.0.0") 
```

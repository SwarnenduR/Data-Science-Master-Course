## Explain why we have to use the Exception class while creating a Custom Exception.

#### Custom exceptions are just regular classes that inherit from the Exception class. This makes it super easy to create our own custom exceptions, which can make our programs easier to follow and more readable. An exception need not be complicated, just inherit from Exception. It’s ok to have a custom Exception subclass that only pass-es - our exception doesn’t need to do anything fancy to be useful. Having custom exceptions - tailored to our specific use cases and that we can raise and catch in specific circumstances - can make our code much more readable and robust, and reduce the amount of code we write later to try and figure out what exactly went wrong.

## Question 3

### What errors are defined in the ArithmeticError class?
#### The arithmetic exception is a type of unusual outcome or unchecked error of the code, which is thrown or raised whenever a wrong mathematical or arithmetic operation appears in the code during run time like: OverflowError, ZeroDivisionError, FloatingPointError

```python
## Example 1 - division by zero
try:
    a = 10
    a / 0
except ArithmeticError as e:
    print(e)
```

```python
## Example 2 - Overflow error
print("Program for showing overflow error\n")
import math
print("The exponential value is")
try:
    math.exp(1000)
except ArithmeticError as e:
    print(e)
```

## Question 4

### Why LookupError class is used?
#### LookupError Exception is the Base class for errors raised when something can't be found. The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid: IndexError, KeyError

```python
## Example for key error
try:
    dict = {1: [2, 3, 4, 5 ], "key": "Swarnendu"}
    dict["key1"]
except KeyError as e:
    print(e)
```

```python
## Example for index error
try:
    l = [1, 2, 3, 4, 5]
    l[10]
except IndexError as e:
    print(e)
```

## Question 5

### Explain ImportError.

#### This error generally occurs when a class cannot be imported due to one of the following reasons: The imported class is in a circular dependency. The imported class is unavailable or was not created. The imported class name is misspelled. The imported class from a module is misplaced. The imported class is unavailable in the Python library.

### What is ModuleNotFoundError?

#### When we try to import a module in a Python file, Python tries to resolve this module in several ways. Sometimes, Python throws the ModuleNotFoundError afterward. As the name implies, this error occurs when we are trying to access or use a module that cannot be found.

### List down some best practices for exception handling in python.

#### 1. Try to use always a specific exception instead of exception super class.
#### 2. Print always a valid message to increae the readability. 
#### 3. Always try to log the errors in a file for permanent use instead just printing the errors in console.
#### 4. Always avoid to write multiple error handling exception for the same exception/at the same time.
#### 5. Prepare proper documentation.
#### 6. Cleanup all the resources after use, for example database connectivity close or network connectivity close or file close.
